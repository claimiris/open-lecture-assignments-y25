Assumption in GLMS:  
Response variable follows a distribution in the exponential family and can be representad as:
$$
f_X(x \mid \theta) = b(x) e^{\left(n^{\top}(\theta) T(x)-a(\theta)) \right.}
$$
  
For Poisson special case:  
The PMF:
$$
\begin{aligned}
f_X(X=x) & =\frac{e^{-\mu} \mu^x}{x!} \quad x=0,1,2,.. \\
& =\frac{1}{x!} e^{\left(xlog(\mu)-a(\mu))\right.}
\end{aligned}
$$
Comparing with general representation:  
Base measure, $b(x)=\frac{1}{x!}$  
Sufficient statistic, $T(x)=x$  
Natural parameter, $\eta(\theta) = \log (\mu)$  
Log partition function, $a(\theta)=e^{\eta(\theta)}=\mu$  
Given that variance exceeds mean,  

$$
\operatorname{Var}(X)>E(X) \rightarrow emperically found
$$

$$
E(X)=a^{\prime}(\theta)=\mu \quad \operatorname{Var}(X)=a^{\prime \prime}(\theta)= \mu
$$

PMF of Negative Binomial(NB) Distribution:  
It model $k$ failures before $r$th success in a sequence of independent Bernoulli distributions.  

$$
\begin{align*}
P(X=k) &= C_k^{k+r-1}p^rq^k, \text { where $p$ is the probability of success}\\
&=C_k^{k+r-1}p^r(1-p)^k \\
&= C_k^{k+r-1} \exp (r \log(p) +k \log (1-p)) \\
\end{align*}
$$
$$
\quad
$$
$$
\begin{aligned}
&b(x=k)= C_k^{k+k-1} \\
& T(x=k)=k \\
& \eta(\theta) = \log (1-p) \\
&( e^n=1-p, p=1-e^n )\\
& a(\theta)=-r \log p=-r \log \left(1-e^n\right) \\
\end{aligned}
$$

$\therefore$ Can be writtin as an exponential family.
$$
\begin{aligned}
E(X)=a^{\prime}(\theta) & =\frac{r e^\eta}{1-e^\eta}=\frac{r(1-p)}{p}=\mu \\
Var(X)=a^{\prime \prime}(\theta) & =r\left(\frac{\left(1-e^\eta\right) e^n+e^\eta\left(e^\eta\right)}{\left(1-e^\eta\right)^2}\right) \\
& =\frac{r e^\eta}{\left(1-e^\eta\right)^2}= \frac{r(1-p)}{p^2}
\end{aligned}
$$
$$
\begin{aligned}
& Var(X)=\mu+\frac{\mu^2}{r} \\
& \mu=\frac{r(1-p)}{r} \\
\end{aligned}
$$
Canonical parameter $\eta=\log (1-p)$ and is not $\log (\mu)$.  
Using $\log$ link to model mean:
$$
\log \mu= X^{\top} \beta
$$

PMF:
$$
P(X=x)=C_x^{x+r-1}\left(\frac{r}{\mu+r}\right)^r\left(\frac{\mu}{\mu+r}\right)^x
$$
$r \rightarrow$ dispension term.  
Poisson forces $Var(X)$=$\mu$. But here, $Var(X)=\mu+\frac{\mu^2}{r}$, which uses $r$ as dispersion term to model any excess variance. Hence, NB Distribution is used in place of Poisson in cases of overdispersion.
$\beta$ is co-efficient vector.  
Whan $r$ is large, NB $\approx$ Poisson distribution. However, this doen't handle underdispersion.

$$
\begin{aligned}

$$

c. False

PCA minimises the mean squared error between data points and their perpendicular projections onto the line.  
Projection:
$$
X_{\text {proj }}=X W\\
X= \text {Data Matrix} \\
W =\text { Matrix set of first few vectors of weights or coefficients from V(from SVD)}
$$
Reconstruction:
$$
X_{\text {recon }}= X_{\text {proj }} W^{\top}
$$  
Error to be minimised:
$$
\frac{1}{N} \sum_{n=1}^N\left\|X_{\text {proj }(n)}-X_{\text {recon }(n)}\right\|^2
$$

In Linear Regression, the squared residuals are minimised (vertical distances between data points and the line).  
Error to be minimised:  
$$
\sum\left(y_{i}-\hat{y}_i\right)^2
$$  
Unless the there is perfect linear relationship between the data, the lines won't natch. The two methods aim to reduce two different errors.

```python
#Q8
image=Image.open("lily.jpg")
image=np.array(image)

def svd_compressor(image, order):

    compressed = np.zeros(image.shape)

    for i in range(3):
        U, S, Vt = np.linalg.svd(image[:,:,i])
        Ui = U[:, :order]
        Vi = Vt[:order, :]
        Si = np.diag(S[:order])
        compressed[:,:,i] = np.dot(Ui, np.dot(Si, Vi))

    compressed = np.clip(compressed, 0, 255).astype(int)

    return compressed

orders = [5,10,20,100]
plt.figure(figsize=(14, 4))

for i in range(len(orders)):

    order = orders[i]
    compressed = svd_compressor(image, order)
    plt.subplot(1, len(orders)+1, i + 1)
    plt.imshow(compressed)
    plt.title(f"Components: {order}")
    plt.axis("off")

plt.subplot(1, len(orders)+1, len(orders)+1)
plt.imshow(image)
plt.title("Original")
plt.axis("off")

plt.suptitle("Compression at different orders")
plt.show()
```
