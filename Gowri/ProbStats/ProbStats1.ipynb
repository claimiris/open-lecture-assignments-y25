##Q1

Given $Y=|Z|$
$$
\begin{align}
& Z \sim N(0,1) \\
& f_z(z)=\frac{1}{\sqrt{2 \pi}} e^{-\frac{z^2}{2}}
\end{align}
$$

Using CDF,
$$
\begin{align}
F_Y(y) & =P(Y \leq y) \\
 &=P(|Z| \leq y) \\
 &=P(-y \leq Z \leq y)  \\
F_Y(y) & =F_z(y)-F_z(-y)
\end{align}
$$

Differentiating both sides with respect to y to get the PDF,
$$
\begin{align}
\frac{d F_Y(y)}{dy} = f_Y(y) \\
f_Y(y) = f_Z(y) + f_Y(-y) \\
\end{align}
$$

$$
\begin{align}
= \frac{2}{\sqrt{2\pi}} e^{-\frac{y^2}{2}} \\
= {\sqrt\frac{2}{\pi}} e^{-\frac{y^2}{2}}, y \geq 0
\end{align}
$$

Finding Expectation,

$$
\begin{align}
E[Y] = \int_{-\infty}^{+\infty} yf_Y(y)dy \
& = \int_{-\infty}^{+\infty} y \sqrt{\frac{2}{\pi}} e^{-\frac{y^2}{2}} dy
\end{align}
$$

Substituting $\frac{y^2}{2} = t$

$$
\frac{2y}{2} dy = dt
$$

$$
\begin{align}
E[Y] = \sqrt{\frac{2}{\pi}} \int_{0}^{\infty} e^{-t} dt
= \sqrt{\frac{2}{\pi}} \left( -e^{-t} \right) \bigg|_{0}^{\infty} = \sqrt{\frac{2}{\pi}}
\end{align}
$$

Finding $E[Y^2]$,
$$
\begin{align}
E[Y^2]=E[Z^2]\
& =\sigma_Z^2 +\mu_z^2 =1
\end{align}
$$

Finding Variance,
$$
\text{Var}(Y) = E[Y^2] - \left( E[Y] \right)^2 \\
= 1 - \frac{2}{\pi}
$$

Finding $E[Y^3]$,
$$
\begin{align}
E[Y^3] = \int_{-\infty}^{+\infty} y^3 \, f_Y(y) \, dy
= \int_{-\infty}^{0} y^3 \times 0  dy + \int_{0}^{\infty} y^3 \sqrt{\frac{2}{\pi}} e^{-\frac{y^2}{2}} \, dy
\end{align}
$$

Substituting  $\frac{y^2}{2} = t$,  
$$
\frac{2y}{2} dy = dt
$$

$$
E[Y^3] = \int_{0}^{\infty} 2t \sqrt{\frac{2}{\pi}} e^{-t} \, dt  = 2 \sqrt{\frac{2}{\pi}} \int_{0}^{\infty} t e^{-t} \, dt
$$

$$
E[Y^3] = 2 \sqrt{\frac{2}{\pi}} \times 1
$$

##Q3

Given,

$$
T_1 \sim \text{Exp}(\lambda_1), \quad T_2 \sim \text{Exp}(\lambda_2)
$$

The exponential distribution probability density function:
$$
f_X(x) =
\begin{cases}
\lambda e^{-\lambda x}, & x > 0, \lambda > 0 \\
0\quad \quad, & \text{otherwise}
\end{cases}
$$

Calculating $P( T_1 \lt T_2 )$,  

Since the random variables are independent,
$$
P(T_1 < T_2) = \int_{0}^{\infty} \int_{t_1}^{\infty} f_{T_1}(t_1) f_{T_2}(t_2) \, dt_2 \, dt_1
$$

$$
= \lambda_1 \lambda_2 \int_{0}^{\infty} \int_{t_1}^{\infty} e^{-\lambda_1 t_1 - \lambda_2 t_2} \, dt_2 \, dt_1
$$

First integrating over $t_2$:

$$
= \lambda_1 \int_{0}^{\infty} e^{-\lambda_1 t_1} \left( \int_{t_1}^{\infty} \lambda_2 e^{-\lambda_2 t_2} \, dt_2 \right) dt_1 \\
= \lambda_1 \int_{0}^{\infty} e^{-\lambda_1 t_1} e^{-\lambda_2 t_1} \, dt_1 \\
= \lambda_1 \int_{0}^{\infty} e^{-(\lambda_1 + \lambda_2) t_1} \, dt_1 \\
= \frac{\lambda_1}{\lambda_1 + \lambda_2}
$$

##Q4
(a)  
Let $X$ be a binary random variable representing the transmitted message:
$$
X = \begin{cases}
1 & \text{if message is } yes \\
0 & \text{if message is } no
\end{cases}
$$

The noisy channel is modeled as additive Gaussian noise:
$Y \sim \mathcal{N}(0, \sigma^2)$

The received signal $Z$ is,
$Z = X + Y$  

The probability of correct interpretation is:
$$
\begin{align}
P(\text{correct}) &= P(Z > \tfrac{1}{2} \mid X = 1)P(X = 1) \\
&\quad + P(Z < \tfrac{1}{2} \mid X = 0)P(X = 0)
\end{align}
$$

$$
\begin{align}
P(\text{correct}) &= P(X = 1)P(Y > -\tfrac{1}{2}) + P(X = 0)P(Y < \tfrac{1}{2}) \\
&= P(X = 1)[1 - \Phi(-\tfrac{1}{2\sigma})] + P(X = 0)\Phi(\tfrac{1}{2\sigma}) \\
&= \Phi(\tfrac{1}{2\sigma})[P(X = 1) + P(X = 0)] \\
&\text{Since it is sure that the message is transmitted, } P(X = 1) + P(X = 0) = 1\\
&= \Phi(\tfrac{1}{2\sigma})
\end{align}
$$
where $\Phi$ is the standard normal CDF.  
Hence, the probability that the received signal falls in the correct decision region is $\Phi(\frac{1}{2\sigma})$  


Behavior of the probability as noise varies extremely:  
Variation in noise can be interpreted using the variance of the distribution.  
As the noise becomes very small,  $\sigma \to 0$:  
$\lim_{\sigma \to 0} \Phi(\frac{1}{2\sigma}) = \Phi(+\infty) = 1$

Intuitively, the probability of correctly interpreting the message tends to 1 as noise becomes negligible.
As noise becomes very large,  $\sigma \rightarrow \infty$:  
$\lim_{\sigma \to \infty} \Phi(\frac{1}{2\sigma}) = \Phi(0) = \frac{1}{2}$

Intuitively, we understand that as $\sigma \rightarrow \infty$ the spread of the data is very large and the actual message and noise conflate into being correctly interpreted only half the time. The probability degrades to $\frac{1}{2}$ .
