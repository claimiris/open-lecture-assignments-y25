#Q1

Bayes' Formula:
$$
P(H)=P(H) \frac{P(e \mid H)}{P(e)}
$$

H - Alex is an expert  
e - Alex hitting 3 bull's eye out of 5  
  
Prior:
$$
f(H)=0.01
$$
  
Event of hitting bull's eyes in trials follows Binomial Distribution:
$$
X \sim \operatorname{Binom}(5, p)
$$

If the is an expert, $p=0.7$
$$
P(e \mid H)=\binom{5}{3}(0.7)^3(0.3)^2
$$

If the is not an expert, $p=0.1$
$$
P(e \mid \bar{H})=\binom{5}{3}(0.1)^3(0.9)^2
$$

By Total probability theorem,
$$
P(e)=P(e \mid H) P(H)+P(e \mid \bar{H}) P(\bar{H})
$$

Posterior:
$$
\begin{aligned}
P(H \mid e) & =\frac{\binom{5}{3} \times(0.7)^3 \times(0.3)^2 \times(0.01)}{\binom{5}{3} \times(0.7)^3 \times(0.3)^2 \times(0.01)+\binom{5}{3} \times(0.1)^3 \times(0.9)^2 \times(0.99)} \\
& =0.278
\end{aligned}
$$

The probability of him being an expert goes from $1 \%$ to $27.8 \%$. This also stamp from understanding that it makes more sense for him to hit 3 bull's eye out of 5 if he is an expert.
  
If prior, $P(H)$ was 0.2 , posterior comes out to be 0.905. The increased belief at the very start contributes to greatly increasing the posterior.

##Q2
For $T$ $\sim \operatorname{Exp}(\lambda)$
  
$$
f_{T}(t)=\left\{\begin{array}{cl}
\lambda e^{-\lambda t}, & t \geqslant 0 \\
0, & t<0
\end{array}\right.
$$

We are considering only $T \geqslant 10$.
Total probability must remain 1.
$$
f_{T \mid T \geqslant 10}(t)= \frac{f_T(t)}{P(T \geqslant 10)}=\lambda e^{-\lambda(t-10)}, t \geqslant 10
$$

Part A:  
Likelihood function:
$$
L(\lambda)=\prod_{i=1}^n \lambda e^{-\lambda\left(t_i-10\right)}=\lambda^n e^{-\lambda\left[\sum_{i=1}^n\left(t_i-10\right)\right]}
$$

Finding MLE using log-likelihood.
$$
\begin{aligned}
& l(\lambda)=\log [L(\lambda)] \\
& l(\lambda)=n \log \lambda-\lambda \sum_{i=1}^n\left(t_i -10\right)
\end{aligned}
$$

Differentiating with respect to $\lambda$,
$$
\frac{d l(\lambda)}{d \lambda} = \frac{n}{\lambda}-{\sum_{i=1}^n\left(t_i-10\right) }
$$

Equating to 0 and finding $\hat{\lambda}$.
$$
\hat{\lambda}=\frac{n}{\sum_{i=1}^n\left(t_i-10\right) }
$$

Not considering truncation would have resulted in using the region with no data points giving a wrong $\lambda$, a cover estimate.  
Missing data points considering the MLE results in a wrong estimate of $\lambda$. This could be rectified by considering the erroneous working of the device in the PDF.


Part B:  
Gamma prior on $\lambda$,
$$
\begin{gathered}
\lambda \sim \operatorname{Gamma}(\alpha, \beta) \\
P(\lambda)=\frac{\beta^\alpha}{\Gamma(\alpha)} \not(\lambda)^{\alpha-1} e^{-\beta \lambda}, \lambda>0 \\
P(\lambda \mid \text { data })=P(\lambda) \times P(\text { data } \mid \lambda)
\end{gathered}
$$

$P(\operatorname{data} \mid \lambda)$ is the likelihood function.

$$
\begin{aligned}
P(\lambda \mid \text {data) }= & \frac{\beta^\alpha}{\Gamma(\alpha)} \lambda^{\alpha-1} e^{-\beta \lambda} \times \lambda^n e^{-\lambda\left[\sum_{i=1}^n(t_i-10)\right]} \\
& =\frac{\beta^\alpha}{\Gamma(\alpha)} \times \lambda^{n+\alpha-1} \times e^{-\lambda\left[\beta+ \sum_{i=1}^n(t_i-10)\right]}
\end{aligned}
$$

Taking $\log$ likelihood and differentiating with respect to $\lambda$ to find $\hat{\lambda}_{\text {MAP }}$.
$$
\hat{\lambda}_{M A P}=\frac{(n+\alpha-1)}{\beta+\sum_{i=1}^n\left(t_i-10\right)}
$$

Prior knowledge sets values for $\alpha$ and $\beta$ which influences the estimate of $\lambda$.  
MAP is preferred over MLE in cases where sample size is small as it incorporates prior information giving a better overall estimate. Additionally, outliers and errors will have a lower impact on the final estimate.
